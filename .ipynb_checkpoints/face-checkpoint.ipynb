```python
import cv2 # opencv
import matplotlib.pyplot as plt # matplotlibの描画系
%matplotlib inline
```

```python
#画像読み込み
img_path = "./o_picts/ims_1/10.jpg" # 画像のパス
```

```python
# cascade の学習結果は https://github.com/Itseez/opencv/tree/master/data/haarcascades から落とせる
cascade_path = "cascade/lbpcascade_animeface.xml"
cascade = cv2.CascadeClassifier(cascade_path) # カスケード分類器を作成
```

```python
#画像読み込み
img_raw = cv2.imread(img_path) # 画像を読み込む

#画像縮小
height = img_raw.shape[0]
width = img_raw.shape[1]
re_width = int(500/height*width)
re_height = 500
img = cv2.resize(img_raw,(re_width,re_height))
```

```python
# opencvの処理系では (B,G,R) で管理されているが、
# matplotlib では (R,G,B) で画像を認識するので、
# 表示用に一旦変換する
show_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
plt.imshow(show_img) #表示
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # グレースケール化
```

```python
#顔認識(各画像スケールにおける縮小量,)
facerect = cascade.detectMultiScale(img_gray, scaleFactor=1.2, minNeighbors=1, minSize=(10,10)) 
```

```python
#結果出力用処理
img_result = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # RGB化
rect_color = (0, 255, 0) # 矩形の色 (B=0, G=255, R=0)
if len(facerect) > 0:
    for rect in facerect:
        # 検出範囲を矩形で囲む
        cv2.rectangle(img_result, tuple(rect[0:2]), tuple(rect[0:2] + rect[2:4]), rect_color, thickness=2)
```

```python
#結果出力
plt.imshow(img_result)
```

```python
print(facerect)
```

```python
#顔だけ切り出して保存
i = 0;
for rect in facerect:
    x = rect[0]
    y = rect[1]
    width = rect[2]
    height = rect[3]    
    #cut
    dst = img[y:y+height, x:x+width]    
    cv2.imwrite(str(i)+".png", dst)
    i += 1
```
